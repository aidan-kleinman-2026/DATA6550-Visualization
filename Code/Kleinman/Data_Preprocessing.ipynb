```python
#Code Block 1
## import libraries

import os
import sys

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_style("whitegrid")

%matplotlib inline
```

```python
#Code Block 2
## load the data

df_train = pd.read_csv("../../data/train.csv")
df_test = pd.read_csv("../../data/test.csv")
df_gender = pd.read_csv("../../data/gender_submission.csv")
```

```python
#Code Block 3
## view the training data

display(df_train.info())
df_train.head()

# PassengerID -> unique identifier for each passenger
# Survived -> target variable (0 = did not survive, 1 = survived)
# Pclass -> passenger class (1 = 1st class, 2 = 2nd class, 3 = 3rd class)
# Name -> name
# Sex -> gender
# Age -> age in years
# SibSp -> number of siblings/spouses aboard the Titanic
# Parch -> number of parents/children aboard the Titanic
# Ticket -> ticket number
# Fare -> passenger fare
# Cabin -> cabin number
# Embarked -> port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)
```

```python
#Code Block 4
## view the test data

display(df_test.info())
df_test.head()
```

```python
#Code Block 5
## view the test submission - just an example we do not need to use this data

display(df_gender.info())
df_gender.head()
```

```python
#Code Block 6
## feature engineering idea - create a new feature called "title" which extracts the title from the name column
## feature engineering idea - create a new feature called "family_size" which is the sum of SibSp and Parch + 1 (for the passenger themselves)
## handle missing values

display(df_train.isnull().sum())
display(df_test.isnull().sum())
```

```python
#Code Block 6
## check the percentage of missing values in each column

display(df_train.isnull().sum()/len(df_train))
display(df_test.isnull().sum()/len(df_test))
```

```python
#Code Block 7
## fill missing values in the fare column with 0 -- snuck onto ship
## fill missing values in the fare column with the median fare for passengers in 3rd class -- most likely the passenger was in 3rd class and did not pay for their ticket
display(df_test[df_test["Fare"].isnull()])
median_fare = df_test[df_test["Pclass"]==3]["Fare"].median()
display(median_fare)
df_test["Fare"].fillna(median_fare, inplace=True)
```

```python
#Code Block 8
## create the "title" feature by extracting the title from the name column
## Mr., Mrs., Miss., Master., etc.
display(df_train["Name"].head())
df_train["Title"] = df_train["Name"].str.extract(" ([A-Za-z]+)\.", expand=False)
display(df_train["Title"].value_counts(dropna = False))

def clean_title(title):
    if title in ["Mr", "Mrs", "Miss", "Master", "Dr", "Rev"]:
        return title
    elif title == "Ms":
        return "Miss"
    elif title == "Sir":
        return "Mr"
    else: return "Other"

df_train["Title"] = df_train["Title"].apply(clean_title)
df_test["Title"] = df_test["Name"].str.extract(" ([A-Za-z]+)\.", expand=False)
df_test["Title"] = df_test["Title"].apply(clean_title)
display(df_train["Title"].value_counts(dropna = False))
display(df_test["Title"].value_counts(dropna = False))
```

```python
#Code Block 9
## fill the missing values for age with the median age for each SibSp and Parch

median_age = df_train.groupby(["SibSp", "Parch"])["Age"].median()
overall_median_age = df_train["Age"].median()
print(overall_median_age)

def fill_age(row):
    if pd.isnull(row["Age"]):
        key = (row["SibSp"], row["Parch"])
        if key in median_age:
            if pd.isnull(median_age[key]):
                return overall_median_age
            else:
                return median_age[key]
        else:
            return overall_median_age
    else:
        return row["Age"]
    
df_train["Age"] = df_train.apply(fill_age, axis=1)
df_test["Age"] = df_test.apply(fill_age, axis = 1)

display(df_train.isnull().sum())
display(df_test.isnull().sum())
```

```python
#Code Block 10
## drop cabin values

df_train.drop("Cabin", axis=1, inplace=True)
df_test.drop("Cabin", axis=1, inplace=True)
```

```python
#Code Block 11
## embarked values
display(df_train[df_train["Embarked"].isnull()])

## fill the missing values with the port that most passengers embarked from
most_common_port = df_train["Embarked"].mode()[0]
print(most_common_port)
df_train["Embarked"] = df_train["Embarked"].fillna(most_common_port)

df_train.isnull().sum()
```

```python
#Code Block 12
## create the "family_size" feature by summing SibSp and Parch + 1 (for the passenger themselves)
df_train["family_size"] = df_train["SibSp"] + df_train["Parch"] + 1
df_test["family_size"] = df_test["SibSp"] + df_test["Parch"] + 1

display(df_train.head(), df_test.head())
```

```python
#Code Block 13
## drop unecessary columns
## name
## ticket

df_train = df_train.drop(["Name", "Ticket"], axis=1)
df_test = df_test.drop(["Name", "Ticket"], axis=1)

display(df_train.info(), df_test.info())
```

```python
#Code Block 14
## save the cleaned data
if not os.path.exists("../../data/cleaned"):
    os.makedirs("../../data/cleaned")
else:
    print("Directory already exists. Saving cleaned data to ../../data/cleaned/cleaned_train.csv and ../../data/cleaned/cleaned_test.csv")

df_train.to_csv("../../data/cleaned/cleaned_train.csv", index=False)
df_test.to_csv("../../data/cleaned/cleaned_test.csv", index=False)
print("Data preprocessing complete. Cleaned data saved to ../../data/cleaned/cleaned_train.csv and ../../data/cleaned/cleaned_test.csv")
```
